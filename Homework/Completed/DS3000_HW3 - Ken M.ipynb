### Question 1 (5pts)
Although you will be scraping the https://boston.craigslist.org URL, your supervisors would like you to write a function that takes two arguments, CL homepage link and a category name, and returns the URL of the category page.

* The idea is that because CL is available across the country, this function could be used to scrape data from CL for other cities as well.
* On CL, rental apartments are posted under "apts / housing" category. If you wanted to scrape data from that page, you would need to access https://boston.craigslist.org/d/apts-housing-for-rent/search/apa
* Because CL uses the same URL structure across all cities in the country, writing a function might be useful!

* Here are some sample function calls:

```python
get_apts_url_from_cl_url("https://boston.craigslist.org", "apts / housing") #url for boston
#this returns the URL of the "apts / housing" page on boston.craigslist.com
```

```python
get_apts_url_from_cl_url("https://newyork.craigslist.org", "apts / housing") #url for new york
#this returns the URL of the "apts / housing" page on newyork.craigslist.com
```

**Hint:**
* your function should take the URL in the first argument
* search for the URL of the second argument on the page provided in the first argument.
* and then return the URL for the category provided in the second argument.

```python
import urllib.request as urllib #gets urllib for pulling url
from bs4 import BeautifulSoup #gets this to pull html

def get_apts_url_from_cl_url(cl_url, cl_label): #makes function for getting link in url
    html_open = urllib.urlopen(cl_url) #opens the url
    html_read = BeautifulSoup(html_open.read(), "lxml") #reads/opens html, "lxml" is for format/html
    html_tags = html_read.find_all("a") #this finds all the html tag [starting] with 'a'
    for tag_item in html_tags: #loops
        if tag_item.get_text().strip() == cl_label: #see if extracted text is [equal to] the label input
            tag_link = tag_item.get('href')
    return cl_url + tag_link #outputs original url with link in addition
    #answer1 goes here
```

### Question 4 (9pts)
Write a code snippet that extracts from your BeautifulSoup object the aforementioned 5 pieces of information about each post: **listing price of the rental apartment**, **neighborhood**, **the title of the post**, **when the listing was posted**, and **the link to the post**.

For each listing, you should store these five values in different columns in the DataFrame.

```python
import time #for pausing/sleeping

titlehd = "result-title hdrlnk" #recurring class-tag used
rental_dict = {} #empty dict.

for column_p in column_names: #goes through each column
    inlist = [] #makes new, empty list, reiterated
    if column_p == "Rent":
        for rent_items in html_soup.find_all("span", class_="result-meta"): #loops all in 'span' tags, that specific <class>
            for rent_p in rent_items.find_all("span", class_="result-price"):
                inlist.append((rent_p.get_text()).strip('$')) #gets rid of $ sign on amounts; adds text to list
            time.sleep(2) #pauses, this to avoid rejection/abortion by host
    elif column_p == "Neighborhood":
        for url_items in html_soup.find_all("a", class_=titlehd):
            urlurl = url_items.get('href') #urlurl is the value in 'href', which is a link
            url_b = urllib.urlopen(urlurl) #opens to urlurl link, for that page's html source
            url_soup = BeautifulSoup(url_b.read(), "lxml")
            for soup_a in url_soup.find_all("meta"): #loop all with meta tag(s)
                if soup_a.get('name') == "geo.placename": #checks if name property in html is equal to geo.placename
                    inlist.append((soup_a.get('content')).title())
            time.sleep(2)
    elif column_p == "Title":
        for title_items in html_soup.find_all("a", class_=titlehd):
            inlist.append(title_items.get_text())
    elif column_p == "Posting Date":
        for time_items in html_soup.find_all("time"):
            full_time = time_items.get('title')
            inlist.append(full_time)
    elif column_p == "URL":
        for url_items in html_soup.find_all("a", class_=titlehd):
            urlurl = url_items.get('href') #get the link
            inlist.append(urlurl) #for each link in loop, add to list
    time.sleep(2)
    rental_dict[column_p.title()] = inlist #adds list to dictionary based on column key

df_rental = pd.DataFrame(rental_dict) #plug-in dictionary to df.

#answer4 goes here
```

### Question 5 (9pts)
Your supervisors would also like you to extract another piece of information from each listing's own page. For each post, they would like you to retrieve the number of bedrooms and bathrooms available in the apartment. This information can be accessed from each listing's own page. You should study that page very carefully!

* Your code snippet should look through the above BeautifulSoup object
* For each listing, go to the listing's page
* Retrieve the bedroom and bathroom information and store it in a list, bed_bath_list
* Before moving on to the next page in your code, you might find it useful to pause the program for 5-10 seconds. CL is really snippy about people crawling their site.

```python
import time

bed_bath_list = [] #empty list

for bos_bb in html_soup.find_all("p"): #goes through html p properties in loop
    for bos_bb_links in bos_bb.find_all("a", class_="result-title hdrlnk"):
        bos_url = bos_bb_links.get('href') #get value in 'href'
        bos_url_sc = urllib.urlopen(bos_url) #use value (link) to be opened
        bos_page = BeautifulSoup(bos_url_sc.read(), "lxml") #read html when open value link
        for page_ii in bos_page.find_all("span", class_="shared-line-bubble"):
            if page_ii.get('data-today_msg') != 'available now': #sees if that property is not equal to "available now"
                if str(page_ii.get_text()).endswith("2") == False: #to skip the value(s) of ft2 (sq. ft.)
                    str_page_ii = str(page_ii.get_text()) #turn html part into string, put into variable
                    bed_bath_list.append(str_page_ii) #add that string (value) to list
            time.sleep(1) #pause for a second
#answer5 goes here
```
