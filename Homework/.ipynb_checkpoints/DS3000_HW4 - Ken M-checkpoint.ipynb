### Question 10 (8pts)
Write a code snippet that groups the universities in the combined dataframe by State. Your code should display a summary dataframe, **grouped_by_state**, containing 
    * the number of universities from each state, 
    * the average number of international students in each state, and 
    * the density of the international students at universities from each state.

Student density will be computed based on the Proportion column:
    * If the average value of the Proportion column for the universities in each state is greater than 50, the density should be "High".
    * If the average is greater than 30, then the density should be "Moderate"
    * If the average is less than 30, the density should be "Low"
    
**Please study the sample output below.**

```python
def student_density (p_item): #created function to see the density based on the proportion(s) of each state
    if (p_item.max() < 30):
        return "Low" #under 30, "Low"
    elif (p_item.max() > 30 and p_item.max() < 50):
        return "Moderate" #between 30 and 50, "Moderate"
    elif (p_item.max() > 50):
        return "High" #over 50, "High"

grouped_by_state = pd.DataFrame() #produce [new] dataframe (grouped_by_state)
grouped_by_state = combined.groupby("State").agg({"State": "count", "Proportion": [student_density], "Students": "mean"})
#have "State" column as index, making states into row, and getting each column "count", "student_density", and "mean"
grouped_by_state.rename(columns={'State':' '}, inplace=True) #"State" should not appear above "count" sub, so removed

#your answer here
```

```python
grouped_by_state #generates the output of the new dataframe, with the headers and subs [of stats] for states
```
