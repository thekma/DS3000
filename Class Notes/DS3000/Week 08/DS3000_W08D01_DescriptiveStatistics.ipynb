## 1. Measures of Central Tendency

```python
import numpy as np
import pandas as pd
```

```python
scores = pd.DataFrame([87, 82, 89, 90, 94, 92, 86, 92, 80, 96], columns=["Scores"])
```

```python
scores
```

### Distribution of the Scores

```python
sorted(scores["Scores"])
```

### 1.1. Mean
* Pandas provides the mean() method
* Can call it on a single column or the entire dataframe

```python
scores["Scores"].mean()
```

### 1.2. Median
* Use median()

```python
scores["Scores"].median()
```

### 1.3. Mode
* Use mode()

```python
scores["Scores"].mode()
```

## 2. Measures of Variability
* Aka measures of dispersion

### 2.1. Range
* No built-in method (no need either)
* Range = max - min

```python
range_scores = scores["Scores"].max() - scores["Scores"].min()
range_scores
```

### 2.2. Variance
* Use var()

```python
scores["Scores"].var()
```

### 2.3. Standard Deviation
* Use std()

```python
scores["Scores"].std()
```

## 4. Normal Distribution

### 4.1. Generating a Random Normal Distribution
* Use the **normal()** method in NumPy's random module
* Accepts three arguments:
    * **loc:** mean of the distribution
    * **scale:** standard deviation of the distribution
    * **size:** number of scores to generate

```python
np.random.seed(1)
s = np.random.normal(loc = 80, scale = 15, size = 10000)
```

```python
s.mean()
```

```python
s.std()
```

### 4.2. Generating a Histogram Using Seaborn
* Can use Seaborn's **distplot()** method

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
import seaborn as sns
sns.distplot(s, kde=False)
```

```python
#Another distribution with the same mean but smaller SD
s2 = np.random.normal(loc = 80, scale = 3, size = 10000)
```

```python
sns.distplot(s2, kde=False)
```

### 4.3. Generating a Histogram Using Plotly
* Use Plotly Express' histogram() method

```python
#let's define a DataFrame for the above distributions
distributions = pd.DataFrame(s, columns = ["D1"])
```

```python
distributions["D2"] = s2
```

```python
distributions.head()
```

```python
import plotly.express as px
fig = px.histogram(distributions, x = "D1")
fig.show()
```

```python
fig = px.histogram(distributions, x = "D2")
fig.show()
```

### 4.4. Generating a Histogram Using Pandas
* Use the built-in hist() method
* Provides support for Seaborn visualizations
* https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.hist.html

```python
distributions.hist(column = "D1")
```

```python
distributions.hist(column="D2", grid=False)
```
